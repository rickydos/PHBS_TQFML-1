```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
data = pd.read_csv('D:/MyDrivers/PYTHON27/final_data_1.csv',sep=",")

data = data.iloc[:,1::]

data_image_array =  np.loadtxt("D:/MyDrivers/PYTHON27/final_data_2.txt")
```

```python
data_train, data_test, image_train, image_test = train_test_split(data,data_image_array,test_size=0.3,random_state=3)

left_eye_x_X_train,left_eye_y_X_train=data_train.loc[:,["sum_left","left_x_L","right_x_L","up_x_L","bottom_x_L","med_x_L"]].values,data_train.loc[:,["sum_left","left_y_L","right_y_L","up_y_L","bottom_y_L","med_y_L"]].values
right_eye_x_X_train,right_eye_y_X_train =data_train.loc[:,["sum_right","left_x_R","right_x_R","up_x_R","bottom_x_R","med_x_R"]].values,data_train.loc[:,["sum_right","left_y_R","right_y_R","up_y_R","bottom_y_R","med_y_R"]].values

left_eye_x_X_test,left_eye_y_X_test=data_test.loc[:,["sum_left","left_x_L","right_x_L","up_x_L","bottom_x_L","med_x_L"]].values,data_test.loc[:,["sum_left","left_y_L","right_y_L","up_y_L","bottom_y_L","med_y_L"]].values
right_eye_x_X_test,right_eye_y_X_test =data_test.loc[:,["sum_right","left_x_R","right_x_R","up_x_R","bottom_x_R","med_x_R"]].values,data_test.loc[:,["sum_right","left_y_R","right_y_R","up_y_R","bottom_y_R","med_y_R"]].values

left_eye_x_y_train=data_train.loc[:,["left_eye_center_x"]].values
left_eye_y_y_train=data_train.loc[:,["left_eye_center_y"]].values
right_eye_x_y_train=data_train.loc[:,["right_eye_center_x"]].values
right_eye_y_y_train=data_train.loc[:,["right_eye_center_y"]].values

left_eye_x_y_test=data_test.loc[:,["left_eye_center_x"]].values
left_eye_y_y_test=data_test.loc[:,["left_eye_center_y"]].values
right_eye_x_y_test=data_test.loc[:,["right_eye_center_x"]].values
right_eye_y_y_test=data_test.loc[:,["right_eye_center_y"]].values



```

```python
##CART Decision Tree
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

## left_eye_x
left_eye_x_tree = DecisionTreeRegressor(max_depth=4,random_state=0)
left_eye_x_tree_fit = left_eye_x_tree.fit(left_eye_x_X_train,left_eye_x_y_train)
left_eye_x_y_train_pred = left_eye_x_tree.predict(left_eye_x_X_train)
left_eye_x_y_test_pred = left_eye_x_tree.predict(left_eye_x_X_test)
print('left_eye_x_tree_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(left_eye_x_y_train, left_eye_x_y_train_pred),
        mean_squared_error(left_eye_x_y_test, left_eye_x_y_test_pred)))
print('left_eye_x_tree_R^2 train: %.3f, test: %.3f' % (
        r2_score(left_eye_x_y_train, left_eye_x_y_train_pred),
        r2_score(left_eye_x_y_test, left_eye_x_y_test_pred)))

## left_eye_y
left_eye_y_tree = DecisionTreeRegressor(max_depth=4,random_state=0)
left_eye_y_tree_fit = left_eye_y_tree.fit(left_eye_y_X_train,left_eye_y_y_train)
left_eye_y_y_train_pred = left_eye_y_tree.predict(left_eye_y_X_train)
left_eye_y_y_test_pred = left_eye_y_tree.predict(left_eye_y_X_test)
print('left_eye_y_tree_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(left_eye_y_y_train, left_eye_y_y_train_pred),
        mean_squared_error(left_eye_y_y_test, left_eye_y_y_test_pred)))
print('left_eye_y_tree_R^2 train: %.3f, test: %.3f' % (
        r2_score(left_eye_y_y_train, left_eye_y_y_train_pred),
        r2_score(left_eye_y_y_test, left_eye_y_y_test_pred)))

## right_eye_x
left_eye_x_tree = DecisionTreeRegressor(max_depth=4,random_state=0)
left_eye_x_tree_fit = left_eye_x_tree.fit(left_eye_x_X_train,left_eye_x_y_train)
left_eye_x_y_train_pred = left_eye_x_tree.predict(left_eye_x_X_train)
left_eye_x_y_test_pred = left_eye_x_tree.predict(left_eye_x_X_test)
print('left_eye_x_tree_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(left_eye_x_y_train, left_eye_x_y_train_pred),
        mean_squared_error(left_eye_x_y_test, left_eye_x_y_test_pred)))
print('left_eye_x_tree_R^2 train: %.3f, test: %.3f' % (
        r2_score(left_eye_x_y_train, left_eye_x_y_train_pred),
        r2_score(left_eye_x_y_test, left_eye_x_y_test_pred)))

right_eye_x_tree = DecisionTreeRegressor(max_depth=4,random_state=0)
right_eye_x_tree_fit = right_eye_x_tree.fit(right_eye_x_X_train,right_eye_x_y_train)
right_eye_x_y_train_pred = right_eye_x_tree.predict(right_eye_x_X_train)
right_eye_x_y_test_pred = right_eye_x_tree.predict(right_eye_x_X_test)
print('right_eye_x_tree_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(right_eye_x_y_train, right_eye_x_y_train_pred),
        mean_squared_error(right_eye_x_y_test, right_eye_x_y_test_pred)))
print('right_eye_x_tree_R^2 train: %.3f, test: %.3f' % (
        r2_score(right_eye_x_y_train, right_eye_x_y_train_pred),
        r2_score(right_eye_x_y_test, right_eye_x_y_test_pred)))
## right_eye_y
right_eye_y_tree = DecisionTreeRegressor(max_depth=4,random_state=0)
right_eye_y_tree_fit =right_eye_y_tree.fit(right_eye_y_X_train,right_eye_y_y_train)
right_eye_y_y_train_pred = right_eye_y_tree.predict(right_eye_y_X_train)
right_eye_y_y_test_pred = right_eye_y_tree.predict(right_eye_y_X_test)
print('right_eye_y_tree_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(right_eye_y_y_train, right_eye_y_y_train_pred),
        mean_squared_error(right_eye_y_y_test, right_eye_y_y_test_pred)))
print('left_eye_y_tree_R^2 train: %.3f, test: %.3f' % (
        r2_score(right_eye_y_y_train, right_eye_y_y_train_pred),
        r2_score(right_eye_y_y_test, right_eye_y_y_test_pred)))





```

```python
## Random Forest
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

## left_eye_x
left_eye_x_forest = RandomForestRegressor(n_estimators=100,max_features="sqrt",random_state=1,)  #n_estimators æŒ‡å†³ç­–æ ‘é¢—æ•°ï¼Œn_jobs æŒ‡CPUå†…æ ¸ä½¿ç”¨æ•°
left_eye_x_forest.fit(left_eye_x_X_train,left_eye_x_y_train)
left_eye_x_y_train_forestpred =left_eye_x_forest.predict(left_eye_x_X_train)
left_eye_x_y_test_forestpred = left_eye_x_forest.predict(left_eye_x_X_test)
print('left_eye_x_forest_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(left_eye_x_y_train, left_eye_x_y_train_forestpred),
        mean_squared_error(left_eye_x_y_test, left_eye_x_y_test_forestpred)))
print('left_eye_x_forest_R^2 train: %.3f, test: %.3f' % (
        r2_score(left_eye_x_y_train, left_eye_x_y_train_forestpred),
        r2_score(left_eye_x_y_test, left_eye_x_y_test_forestpred)))

## left_eye_y
left_eye_y_forest = RandomForestRegressor(n_estimators=100,max_features="sqrt",random_state=1,)  #n_estimators æŒ‡å†³ç­–æ ‘é¢—æ•°ï¼Œn_jobs æŒ‡CPUå†…æ ¸ä½¿ç”¨æ•°
left_eye_y_forest.fit(left_eye_y_X_train,left_eye_y_y_train)
left_eye_y_y_train_forestpred =left_eye_y_forest.predict(left_eye_y_X_train)
left_eye_y_y_test_forestpred = left_eye_y_forest.predict(left_eye_y_X_test)
print('left_eye_y_forest_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(left_eye_y_y_train, left_eye_y_y_train_forestpred),
        mean_squared_error(left_eye_y_y_test, left_eye_y_y_test_forestpred)))
print('left_eye_y_forest_R^2 train: %.3f, test: %.3f' % (
        r2_score(left_eye_y_y_train, left_eye_y_y_train_forestpred),
        r2_score(left_eye_y_y_test, left_eye_y_y_test_forestpred)))

## right_eye_x
right_eye_x_forest = RandomForestRegressor(n_estimators=100,max_features="sqrt",random_state=1,)  #n_estimators æŒ‡å†³ç­–æ ‘é¢—æ•°ï¼Œn_jobs æŒ‡CPUå†…æ ¸ä½¿ç”¨æ•°
right_eye_x_forest.fit(right_eye_x_X_train,right_eye_x_y_train)
right_eye_x_y_train_forestpred =right_eye_x_forest.predict(right_eye_x_X_train)
right_eye_x_y_test_forestpred = right_eye_x_forest.predict(right_eye_x_X_test)
print('right_eye_x_forest_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(right_eye_x_y_train, right_eye_x_y_train_forestpred),
        mean_squared_error(right_eye_x_y_test, right_eye_x_y_test_forestpred)))
print('right_eye_x_forest_R^2 train: %.3f, test: %.3f' % (
        r2_score(right_eye_x_y_train, right_eye_x_y_train_forestpred),
        r2_score(right_eye_x_y_test, right_eye_x_y_test_forestpred)))

## right_eye_y
right_eye_y_forest = RandomForestRegressor(n_estimators=100,max_features="sqrt",random_state=1,)  #n_estimators æŒ‡å†³ç­–æ ‘é¢—æ•°ï¼Œn_jobs æŒ‡CPUå†…æ ¸ä½¿ç”¨æ•°
right_eye_y_forest.fit(right_eye_y_X_train,right_eye_y_y_train)
right_eye_y_y_train_forestpred =right_eye_y_forest.predict(right_eye_y_X_train)
right_eye_y_y_test_forestpred = right_eye_y_forest.predict(right_eye_y_X_test)
print('right_eye_y_forest_MSE train: %.3f, test: %.3f' % (
        mean_squared_error(right_eye_y_y_train, right_eye_y_y_train_forestpred),
        mean_squared_error(right_eye_y_y_test, right_eye_y_y_test_forestpred)))
print('right_eye_y_forest_R^2 train: %.3f, test: %.3f' % (
        r2_score(right_eye_y_y_train, right_eye_y_y_train_forestpred),
        r2_score(right_eye_y_y_test, right_eye_y_y_test_forestpred)))

```
